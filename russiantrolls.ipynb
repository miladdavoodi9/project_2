```python
import pandas as pd
from sqlalchemy import create_engine



# bring in the csv files / data
# Source: Kaggle

Russian_Troll_file = "Resource/IRAhandle_tweets_1.csv"
Russian_Troll_df = pd.read_csv(Russian_Troll_file)
Russian_Troll_df.head()

Russian_Troll_file_2 = "Resource/IRAhandle_tweets_2.csv"
Russian_Troll_file_2_df = pd.read_csv(Russian_Troll_file_2)

# Combine dataframes
russian_troll_combined = pd.concat([Russian_Troll_df, Russian_Troll_file_2_df])

russian_troll_combined.head()
```

```python
# ira_tweets 1GB file. Very Large
# Source: published by Twitter

ira_tweets = "Resource/ira_tweets_csv_hashed.csv"
ira_tweets_df = pd.read_csv(ira_tweets)
ira_tweets_df.head()
```

```python
# ir_users file 
# Source: shared by Twitter

ira_users = "Resource/ira_users_csv_hashed.csv"
ira_users_df = pd.read_csv(ira_users)
ira_users_df.head(5)
```

```python
# Drop unnecessary columns in russion_troll_combined df

russian_troll_combined = russian_troll_combined.drop(columns = ['author', 
                                                                'harvested_date', 
                                                                'updates', 
                                                                'account_type', 
                                                                'new_june_2018', 
                                                                'account_category'])
```

```python
# Drop duplicates  
ira_users_df.drop_duplicates("userid", inplace=True)
ira_tweets_df.drop_duplicates("userid", inplace = True)

russian_troll_combined.drop_duplicates('external_author_id', inplace = True)
```

```python
# Set index 
ira_users_df.set_index("userid", inplace = True)
ira_tweets_df.set_index ("userid", inplace = True)

russian_troll_combined.set_index ("external_author_id", inplace = True)
```

```python
# Create connection to postgres
connection_string = "postgres:previouslythepasswordwashere@localhost:5432/Russian Trolls"
engine = create_engine(f'postgresql://{connection_string}')
```

```python
# Confirm connection
engine.table_names()
```

```python
# Send data to PostgresSQL

ira_tweets_df.to_sql(name='ira_tweets', con=engine, if_exists='append', index=True)

ira_users_df.to_sql(name='ira_user', con=engine, if_exists='append', index=True)

russian_troll_combined.to_sql(name='russian_trolls', con=engine, if_exists='append', index=True)
```
